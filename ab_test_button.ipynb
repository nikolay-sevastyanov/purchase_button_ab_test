```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
```

```python
df = pd.read_csv('ab_test_button.csv', delimiter=',')
df.drop(df.columns[0], axis=1, inplace=True)
print(df.head(5))
```

```python
print(df['group'].unique())
```

```python
control = df[df['group']=='control']
test = df[df['group'] == 'test']
```

```python
print(control.head())
```

```python
print(test.head())
```

```python
control_cr = len(control[control['converted']==1])/len(control)
test_cr = len(test[test['converted']==1])/len(test)

cr_lift = test_cr - control_cr

print(f"control_cr = {control_cr:f}")
print(f"test_cr = {test_cr:f}")
print(f"cr_lift = {cr_lift:f}")
```

```python
count = [  len(test[test['converted']==1]), len(control[control['converted']==1]) ]
nobs = [ len(test['converted']), len(control['converted']) ]

z_stat, p_value = proportions_ztest(count, nobs)

print(f"Успехов: {count}")
print(f"Наблюдений: {nobs} \n")

print(f"Z-статистика: {z_stat:.2f}")
print(f"P-value: {p_value} или {p_value*10**158:.2f}/(10^158)")
```
